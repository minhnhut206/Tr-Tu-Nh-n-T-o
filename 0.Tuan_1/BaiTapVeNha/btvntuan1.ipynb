```python
def tron_mang(a, b):
    c = []
    min_len = min(len(a), len(b))
    for i in range(min_len):
        c.append(min(a[i], b[i]))

    c.extend(a[min_len:])
    c.extend(b[min_len:])
    return c
#===========================
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))

```

```python
#a)
def tong_ptu_tamgiactren(a):
  tong = 0
  n = len(a)
  for i in range(n):
    for j in range(n):
      if i+j <= n-1:
        tong+=a[i][j]
  return tong
#b)
def doi_ptuam_thanhduong(a):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if a[i][j] < 0:
        a[i][j] = abs(a[i][j])
#c)
def thay_ptuchan_thanhx(a, x):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if a[i][j] % 2 == 0:
        a[i][j] = x
#d)
def kt_toanchan(a):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if a[i][j]%2!=0:
        print("mảng không toàn ptu chẵn")
        return
  print("mảng toàn ptu chẵn")
  return
#e)
def kt_mang_doixung(a):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if a[i][j] != a[j][i]:
        print("mảng ko đối xứng")
        return
  print("mảng không đối xứng")
#f)
def kt_duongcheochinh_tangdan(a):
  n = len(a)
  for i in range(n - 1):
    if a[i][i] <= a[i+1][i+1]:
      print("mảng ko có đcc tăng dần")
      return
  print("mảng có đcc tăng dần")
#g)
def xuat_tamgiacduoi_duongcheophu(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()
#h)
def duongcheophu_giamdan(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n-1-i] >= a[i+1][n-2-i]:
            print("mảng ko có đcp giảm dần")
            return
    print("mảng có đcp giảm dần")
#================================================
a = [
    [3, -2, 5],
    [4,  6, -1],
    [7,  8,  2]
]
print("Tổng tam giác trên của đường chéo phụ (kể cả chéo phụ:", tong_ptu_tamgiactren(a))
doi_ptuam_thanhduong(a)
print("Sau khi đổi âm:", a)
thay_ptuchan_thanhx(a, 99)
print("Sau khi thay chẵn:", a)
kt_toanchan(a)
kt_mang_doixung(a)
kt_duongcheochinh_tangdan(a)

print("Tam giác dưới chéo phụ:")
xuat_tamgiacduoi_duongcheophu(a)
duongcheophu_giamdan(a)


```

**Pandas** là thư viện dùng để: *xử lý dữ liệu dạng bảng*, *đọc file*, *lọc*, *tính toán* và *thống kê dữ liệu*.

---

**Pandas** có 2 cấu trúc chính như sau:

Series: 1 cột

DataFrame: bảng nhiều cột




ví dụ:

```python
import pandas as pd

data = {
    "ten": ["Nhut", "An", "Binh"],
    "tuoi": [20, 21, 19]
}

df = pd.DataFrame(data)
print(df)
print()
print(df[df["tuoi"] > 20])
print()
print(df["tuoi"].mean())
print()
print(df["tuoi"].max())
```

**Matplotlib** là thư viện dùng để: *Vẽ biểu đồ* (chart) và *trực quan hóa dữ liệu*


ví dụ vẽ biểu đồ đường:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title("Bieu do duong")
plt.xlabel("Truc X")
plt.ylabel("Truc Y")
plt.show()
```

Ví dụ vẽ biểu đồ cột:

```python
plt.bar(["A", "B", "C"], [10, 20, 15])
plt.show()
```

ví dụ vẽ biểu đồ tròn:

```python
plt.pie([30, 40, 30], labels=["A", "B", "C"])
plt.show()
```
