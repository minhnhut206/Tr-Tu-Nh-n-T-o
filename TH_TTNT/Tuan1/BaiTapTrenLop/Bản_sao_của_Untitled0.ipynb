```python
import random
# a)
def tao_mang(m,n):
  a = []
  for i in range(m):
    dong = []
    for j in range(n):
      dong.append(random.randint(-50, 50))
      a.append(dong)
  return a
def xuat_mang(a):
  for i in a:
    print(*i)
# b)
def xuat_dong_k(a, k):
  print("Dòng", k, ": ", *a[k])
# c)
def xuat_cot_k(a, k):
  print("Cột", k, ": ")
  for i in range(len(a)):
    print(a[i][k], end=" ")
  print()
# d)
def tim_dong_tong_max(a):
    max_sum = float('-inf')
    index = -1

    for i in range(len(a)):
        s = sum(a[i])
        if s > max_sum:
            max_sum = s
            index = i
    return index
# e)
def tim_cot_tich_min(a):
  m = len(a)
  n = len(a[0])
  min_tich = float('inf')
  index = -1
  for j in range(n):
    tich = 1
    for i in range(m):
      tich *= a[i][j]
    if min_tich > tich:
      min_tich = tich
      index = j
  return index
# f)
def xuat_dongchan_cotle(a):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if i % 2 == 0 and j % 2 != 0:
        print(a[i][j], end=" ")
  print()
# g)
def tbcong_ptuchan_dongle(a):
  tong = 0
  dem = 0
  for i in range(len(a)):
    if i % 2 != 0:
      for j in range(len(a[0])):
        if a[i][j] % 2 == 0:
          tong += a[i][j]
          dem += 1
  return tong / dem if dem != 0 else 0
# h)
def tbcong_ptubien(a):
  m = len(a)
  n = len(a[0])
  tong = 0
  dem = 0

  for i in range(m):
    for j in range(n):
      if i == 0 or i == m-1 or j == 0 or j == n-1:
        tong += a[i][j]
        dem += 1
  return tong / dem
# i)
def tbtich_ptukobien(a):
  m = len(a)
  n = len(a[0])

  tich = 1
  dem = 0

  for i in range(1, m-1):
    for j in range(1, n-1):
      tich += a[i][j]
      dem += 1
  return tich / dem if dem != 0 else 0
#---------------------------------------------
m, n = 4, 5
a = tao_mang(m, n)
print("mảng 2 chiều a:")
xuat_mang(a)
print("dòng k: ")
xuat_dong_k(a, 1)
print("cột k: ")
xuat_cot_k(a, 2)
print("tổng max= ", tim_dong_tong_max(a))
print("cột tích nhỏ nhất:", tim_cot_tich_min(a))
print("PTu dòng chẵn cột lẽ: ")
xuat_dongchan_cotle(a)
print("trung bình cộng ptu chẵn cột lẽ:")
tbcong_ptuchan_dongle(a)
print("trung bình cộng ptu ở biên:")
tbcong_ptubien(a)
print("trung bình tích ptu không biên:")
tbtich_ptukobien(a)





```

```python
# a)
class SINHVIEN:
  def __init__(self, masv, ten, namsinh, dtb):
    if len(masv) != 10:
      raise ValueError("Mã sinh viên không được quá 10 kí tự!!!")
    if len(ten) > 20:
      raise ValueError("Tối đa 20 ký tự!!!")
    self.masv = masv
    self.ten = ten
    self.namsinh = namsinh
    self.dtb = dtb

def xuat(self):
    print(f"Mã SV: {self.masv}")
    print(f"Tên: {self.ten}")
    print(f"Năm sinh: {self.namsinh}")
    print(f"Điểm TB: {self.dtb}")
# b)
def tao_mang(n):
  dssv = []
  for i in range(n):
    print(f"\n Nhập sinh viên thứ ", {i+1}, ": ")
    masv = input("nhập mã sinh viên: ")
    ten = input("nhập tên sinh viên: ")
    nam = int(input("nhập năm sinh: "))
    dtb = float(input("nhập điểm tb: "))

    sv = SINHVIEN(masv, ten, nam, dtb)
    dssv.append(sv)
  return dssv
# c)
def dem_sv_lenlop(ds):
  dem = 0
  for sv in ds:
    if sv.dtb >= 5:
      dem += 1
  return dem
# d)
def xuat_sv_du20tuoi(ds):
  nam_hientai = 2026
  for sv in ds:
    if nam_hientai - sv.namsinh >= 20:
      sv.xuat()
# e)
def dem_svDH(ds):
  dem = 0
  for sv in ds:
    if sv.masv[2:4] == "DH":
      dem += 1
  return dem
# f)
def dem_svLan(ds):
  dem = 0
  for sv in ds:
    ten = sv.ten.strip().split()[-1]
    if ten == "Lan":
      dem += 1
  return dem
# g)
def dem_sv_hoPhan(ds):
  dem = 0
  for sv in ds:
    ho = sv.ten.strip().split()[0]
    if ho == "Phan":
      dem += 1
  return dem
#==========================================#

n = int(input("Nhập số sinh viên: "))
ds = tao_mang(n)

print("Số SV đủ điều kiện lên lớp:", dem_sv_lenlop(ds))
print("Số SV hệ ĐH:", dem_svDH(ds))
print("Số SV tên Lan:", dem_svLan(ds))
print("Số SV họ Phan:", dem_sv_hoPhan(ds))



```
